Lab 2 Pet store
The Pet store hired us to create an ER diagram to make sense of all the data that their store collects and create a loyalty program for them as well. 
The Loyalty program is connected to all the other purchase tables such as the food purchases and animal purchases. The loyalty program includes the name of the customer, their phone number, and their address. The food purchases encompass the food purchased by the customer and food purchased by the pet store itself. The Purchases table is connected to the pets and fish tables because the food needs to be purchased for the animals by the store. 
When it comes to the inner workings of the pet store we split up the living environments for the animals into Enclosures and aquariums. We further split these up into the specific enclosure or aquarium for the Fish and animals. 
We split up both types of animals (fish and pets) into different species with each having their own section in enclosures (for pets) and aquariums (for fish). This allows the customers to choose what kind of fish or pet that they would like to adopt. We used the “many to many”-type relationship to signify that there are many locations for many different types of fish and types of pets. 
Entities for the project include: locations, Pets, Fish, Aquariums, Enclosures, Food Purchases, Purchases, Loyalty Program, and Animals. Attributes include: the species of pets and fish, the different enclosures and aquariums, store purchases, and customer purchases. 
We chose to have our primary key be the loyalty program. This gives customers different perks for shopping with our store. This primary key would be a card that the customer would scan before they pay for their order. Secondary keys would be the phone number and name of the customer. This would allow them to access their program even if they forgot their card.
